### Gate Cutting to Reduce Circuit Width
- Gate Cutting to Reduce Circuit Width:\
Circuit cutting to reduce the number of qubits in a circuit. We will cut gates to enable us to reconstruct the expectation value of a 4 qubit circuit to a 2 qubit experiments

```python
# Getting library
import matplotlib.pyplot as plt
from qiskit.circuit.library import EfficientSU2
#Efficientsu2 a class within the Qiskit Circuit Library.  
```

###### EfficientSU2 is a parameterized quantum circuit ansatz in Qiskit, designed for variational quantum algorithms (VQAs). It's a popular choice due to its efficiency and ability to generate a wide range of quantum states with a relatively small number of parameters.
Here's a breakdown of its key features and characteristics:
Alternating Layers: EfficientSU2 consists of alternating layers of single-qubit rotations and two-qubit entangling gates.
Single-Qubit Rotations: Each qubit is acted upon by single-qubit rotations, typically rotations around the X and Z axes (Rx and Rz). These rotations introduce the variational parameters that are optimized during the VQA process.   
Two-Qubit Entangling Gates: The entangling gates, usually CNOT (CX) gates, create entanglement between the qubits. The entanglement pattern can be controlled (e.g., linear, circular, full).   
Repetitions (Reps): The number of repetitions (layers) of the single-qubit rotations and entangling gates can be adjusted. More repetitions increase the circuit's expressiveness but also increase the number of parameters and the circuit depth.
Parameterization: The rotation angles in the single-qubit rotations are the variational parameters. These parameters are adjusted during the optimization process to minimize a cost function.
Efficiency: The name "EfficientSU2" comes from its relatively efficient use of quantum gates to generate a diverse set of quantum states.

```python
qc = EfficientSU2( 4, entanglement ="linear", reps = 2).decompose()
# Decompose() expands it into the actual sequence of single-qubit rotations and two-qubit entangling gates that make up the circuit and it's needed before assigning parameters or executing the circuit on a backend.
qc.assign_parameters([0.4] * len( qc.parameters), inplace= True)
# Assigns the numerical value 0.4 to all the parameters in the circuit.
# inplace=True; This argument modifies the circuit qc directly, rather than returning a new circuit.
qc.draw("mpl", filename='c1', scale= 0.8 )
```

##### Specify an observable

```python
from qiskit.quantum_info import SparsePauliOp
#SparsePauliOp class.This class allows to represent observables as a linear combination of Pauli strings
observable = SparsePauliOp(["ZZII", "IZZI", "-IIZZ", "XIXI","ZIZZ", "IXIX"])
# This is a list of Pauli strings. Each string represents a tensor product of Pauli operators (I, X, Y, Z) acting on different qubits.
```

## Step 1: Map; Creating a circuit to cut

```python
from qiskit_addon_cutting import partition_problem

partitioned_problem = partition_problem(
    circuit=qc, partition_labels="AABB", observables=observable.paulis
)
subcircuits = partitioned_problem.subcircuits
subobservables = partitioned_problem.subobservables
bases = partitioned_problem.bases
```
